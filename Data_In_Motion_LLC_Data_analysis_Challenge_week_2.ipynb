```python
# Loading Essential Libraries
import pandas as pd
import numpy as np
from siuba import *
```

```python
# Loading Dataset from URL
euro = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv', sep=',')
```

```python
euro.describe()
```

```python
# How many teams participated in the Euro2012?
euro >> _.Team.count()
```

```python
# What is the number of columns in the dataset?
euro.shape[1]
```

```python
# View only the columns Team, Yellow Cards and Red Cards and assign them to a dataframe called discipline.
discipline = euro >> select('Team', 'Yellow Cards', 'Red Cards')
```

```python
discipline.head()
```

```python
discipline >> arrange(_.get('Red Cards'), _.get('Yellow Cards'))
```

```python
# Calculate the mean Yellow Cards given per Team.
discipline >> group_by(_.Team) >> summarize(mean_yellow_cards=_.get('Yellow Cards').mean())
```

```python
# Filter teams that scored more than 6 goals
euro >> filter(_.Goals > 6) >> select('Team', 'Goals')
```

```python
# Select the teams that start with the letter G.
euro >> filter(_.Team.str.startswith("G")) >> select('Team')
```

```python
# Select the first 7 columns
euro >> select(_[0:7])
```

```python
# Present only the Shooting Accuracy from England, Italy and Russia. in siuba
euro >> filter(_.Team.isin(['England', 'Italy', 'Russia'])) >> select('Shooting Accuracy')
```

```python
# Select all columns except the last 3
euro >> select(_[32:0])
```
