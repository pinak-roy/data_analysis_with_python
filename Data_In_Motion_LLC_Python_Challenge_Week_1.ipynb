```python
# Which was the most-ordered item?
chipo >> count(_.item_name, sort = True) >> select(_.item_name) >> head(1)
```

```python
# For the most-ordered item, how many items were ordered?
chipo >> count(_.item_name, sort = True) >>select(_.n) >> head(1)
```

```python
# What was the most ordered item in the choice_description column?
chipo >> count(_.choice_description, sort=True) >> head(1)
```

```python
# How many items were ordered in total?
chipo >> _.quantity.sum()
```

```python
# Turn the item price into a float
chipo = (chipo >>
mutate(item_price_float=_.item_price.str.replace('$', '', regex=True).astype(float)))
```

```python
# Cheking the new dataset
chipo.head(10)
```

```python
# Adding new column named "Sales Revenue"
chipo["sales_revenue"] = chipo["quantity"]*chipo["item_price_float"]
```

```python
# Cheking Data
chipo.head(10)
```

```python
# How much was the revenue for the period in the dataset?
total_revenue = (chipo >> _.sales_revenue.sum())
```

```python
print(total_revenue)
```

```python
# How many orders were made in the period?
number_of_orders = (chipo >> _.order_id.nunique())
```

```python
print(number_of_orders)
```

```python
# What is the average revenue amount per order?
average_revenue = total_revenue/number_of_orders
```

```python
print(average_revenue)
```
