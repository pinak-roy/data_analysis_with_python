Importing Dataset

Delete the columns ‘Unnamed: 0’ and ‘Id’.

```python
baby = baby >> rename(uno='Unnamed: 0')
```

```python
baby = baby >> select(~_.uno, ~_.Id)
```

```python
baby.head()
```

```python
names = baby >> count(_.Name, sort = True)
```

```python
print(names)
```

How many different names exist in the dataset?

```python
names >> _.Name.count()
```

```python
names >> head(1)
```

What is the standard deviation of count of names?

```python
names >> _.n.std()
```
